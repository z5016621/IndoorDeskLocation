What is [Template Matching?](https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_imgproc/py_template_matching/py_template_matching.html)

Searching and finding the location of a template image in a larger image


The idea here is to find identical regions of an image that match a template we provide, giving a threshold. 
The threshold depends on the accuracy with which we want to detect the template in the source image. 
In this example, a threshold of 0.98 means we want a near perfect match. Near identical thresholds need to be set higher than 0.8


```python
def templateMatch(img, sub_img, threshold=0.98):
    width, height = sub_img.shape[::-1]

    result = cv2.matchTemplate(img,
                               sub_img,
                               cv2.TM_CCOEFF_NORMED)
    
    
    # This particular result is using the CCOEFF NORMED matching method which is a particular operation/algorithym/mathematical 
    # equation in template matching
    
    loc = np.where(result >= threshold)
 
    # Location using numpy where the returned result of the match is equal to or above 0.98

    count = 0
    
    # White - setting the rgb value for white without more numbers to complicate the script 
    # count - number of times the loop is happening for the number of points in zip

    
    print ("Location Information:", loc)

    output_image = img.copy()
    
    #Creating a copy of the output image so the result is not overiding on the original image each time
    
    for pt in zip(*loc):
        # This reformats the data so that the x and y coordinates are paired up, as opposed to being in separate lists  
        
        # Python: cv.Rectangle(img, pt1, pt2, color, thickness=1, lineType=8, shift=0) → None
    
        cv2.rectangle(output_image, 
                        pt,
                        (pt[0] + width, pt[1] + height),
                        WHITE, 
                        thickness=2
                        )
        count +=1
    
    #Adding 1 everytime the loop happens, in this case only happens once

    show_image(output_image, name="test")
    print(count)
    
    
    
print ("This particular example doesn't work because the search image"
       " is not identical to the super image (shown), and therefore returns a result of 0 as there are no matching points (shown in the array brackets).")
templateMatch(super_grey, search_grey, 0.8)

print "----------------------------------------------\n\n\n"

print "This image is a defined area cropped (crop_img) from the original image which will be used to search in the super image"
crop_img = super_grey[200:600, 200:600] # [y:y+h, x:x+w]
show_image(crop_img, name="test")

print "----------------------------------------------\n\n\n"

print "Resulting image showing the correct location of the cropped image over the super image. This is a successful match with 3 points"
templateMatch(super_grey, crop_img, 0.8)

print "----------------------------------------------\n\n\n"

#dtype is data type - int64 = Integer (-9223372036854775808 to 9223372036854775807)
```

What are [KeyPoints](https://stackoverflow.com/questions/29133085/what-are-keypoints-in-image-processing)

They are spatial locations, or points in the image that define what is interesting or what stand out in the image. The reason why keypoints are special is because no matter how the image changes... whether the image rotates, shrinks/expands, is translated or is subject to distortion (i.e. a projective transformation or homography), you should be able to find the same keypoints in this modified image when comparing with the original image.


```python
def onlyMatches(matches,      # Keypoint/Descriptor? matches between the two images
                super_image,  # Is the original overall image
                search_image, # Is the initial image to be found
                imgKeyPoints, # Are the points that stand out in the super image
                subKeyPoints  # Are the points that stand out in the search image
               ):

    #Changing the number in the [] brackets will change the number of matches shown

    usedMatches = matches[:10] # a copy of the whole array
    #colours = [lambda x: (x.distance*2.5, 0, 0) for i in usedMatches]

    finalImg = cv2.drawMatches(super_image, 
                               imgKeyPoints, 
                               search_image,
                               subKeyPoints, 
                               usedMatches, 
                               None, # Output image - doesn't seem to do much
                               matchColor=YELLOW, 
                               singlePointColor=BLUE, 
                               flags=2
                              )

    show_image(finalImg, name="label me soon final img")
```

```python
# RANSAC STUFF
def ransac(matches,
           super_image,
           search_image,
           imageKeyPoints,
           subKeyPoints,           
           RANSACthreshold,
           imageName=""):
    
    show_image(super_image, name="Super Image")
    show_image(search_image)
    
    sourceKeyPointLocations = []
    destinationKeyPointLocations = []
    
    for match in matches:
        sourcePointID = match.queryIdx
        sourceKeyPoint = imageKeyPoints[sourcePointID]
        sourceLocation = sourceKeyPoint.pt
        sourceKeyPointLocations.append(sourceLocation)
        
        destinationPointID = match.trainIdx
        destinationKeyPoint = subKeyPoints[destinationPointID]
        destinationLocation = destinationKeyPoint.pt
        destinationKeyPointLocations.append(destinationLocation)
    
    sourcePointsUnformatted = np.float32(sourceKeyPointLocations)
    destinationPointsUnformatted = np.float32(destinationKeyPointLocations)
    
    sourcePoints = sourcePointsUnformatted.reshape(-1, 1, 2)
    destinationPoints = destinationPointsUnformatted.reshape(-1, 1, 2)
    
    # Converting the keypoints from the list of matches to an arguement for findHomography         

    M, mask = cv2.findHomography(sourcePoints, destinationPoints, cv2.RANSAC, RANSACthreshold)

    # Generating the original search image size and placing it on the image starting at the coordinates 0,0
    height, width = search_image.shape[:2]
    points = np.float32([[0,         0         ], 
                         [0,         height - 1], 
                         [width - 1, height - 1], 
                         [width - 1, 0         ]]).reshape(-1, 1, 2)

    #import pdb; pdb.set_trace()

    #invert matrix - search image transformation initially along the wrong path
    M = np.linalg.inv(M)

    #Warp corners appropriately - moving the initial search image into the correct position
    destination = cv2.perspectiveTransform(points, M)
    #destination = points

    output_image = super_image.copy()
    #Draw the sides of the rectangle
    img2 = cv2.polylines(output_image, [np.int32(destination)], True, RED, 3, cv2.LINE_AA)

    show_image(output_image, name="Result")
```

Default Features created within the ORB_create Object [ORB](https://docs.opencv.org/master/db/d95/classcv_1_1ORB.html#aeff0cbe668659b7ca14bb85ff1c4073b)

Keypoints are the blue circles which appear on the images
Descriptors are primarily concerned with both the scale and the orientation of the keypoint

The circle correlates to the scale the point was detected at. The process detects keypoints at different scales and then they are all combined to create the final output. The larger the circle, the larger the scale was that the point was detected at

The line fron the centre of the circle dictates the orientation of the keypoint. The orientation is determined by analysing the surrounding pixels to determine how that pixel fits best. 


```python
def featureMatch(super_image, search_image, RANSACthreshold=5, colour=RED, do_RANSAC=False, imageName="test"):
    
    orb = cv2.ORB_create()      #Creating an ORB object which will initiate an ORB detector
    
    imgKeyPoints, imgDescriptors = orb.detectAndCompute(super_image, None)
    subKeyPoints, subDescriptors = orb.detectAndCompute(search_image, None)


    #Is finding the keypoints and the descriptors of each of the images, then computing the BRIEF descriptors with the ORB

    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    
    #Creating the BF Matcher object
    #cv2.Norm_Hamming, used for binary string based descriptors like ORB which uses Hamming distance as measurement
    #Hamming Distance  between two strings of equal length is the number of positions at which the corresponding 
    #symbols are different
    #crossCheck is set to true, given each match of features found, each feature must be the other's best match from that image

    matches = bf.match(imgDescriptors, subDescriptors)
    
    def findMatchDistance(match):
        """Finds the distance value of a match for the sorting function below"""
        return match.distance

    matches = sorted(matches, key = findMatchDistance)
    
    print "This is the distance of the first match: " + str(findMatchDistance(matches[0]))
    
    # First line is matching the descriptors in both of the images
    # Second line is saying sort the descriptors in order of their distance 
    # Distance describes how wgood the match is

    
    #import pdb; pdb.set_trace()
       
    if do_RANSAC:
        ransac(matches,       # Matches are the descriptors from both images that match starting from the shortest distance which is the most accurate
               super_image,
               search_image,
               imgKeyPoints,
               subKeyPoints,           
               RANSACthreshold,
               imageName)
        print "Result using the RANSAC method"
        
    else:
        print "Result using only matches"
        onlyMatches(matches,
                    super_image,
                    search_image,
                    imgKeyPoints,
                    subKeyPoints)
        
    print "----------------------------------------------\n\n"


featureMatch(super_rgb,  search_rgb, RANSACthreshold=30, do_RANSAC=True, imageName="test")
featureMatch(super_rgb,  search_rgb, RANSACthreshold=30, do_RANSAC=False, imageName="test")

featureMatch(super_grey, crop_img,   RANSACthreshold=30, do_RANSAC=True, imageName="")
featureMatch(super_grey, crop_img,   RANSACthreshold=30, do_RANSAC=False, imageName="")
```
